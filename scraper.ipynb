```python
from bs4 import BeautifulSoup
import requests
```

```python
def scrape_one(bird_id):
    url = 'https://ebird.org/species/' + bird_id
    html = BeautifulSoup(requests.get(url).text, 'html')
    return html.find('span', class_='Heading-main Media--hero-title').get_text()
```

```python
scrape_one('higtin3')
```
